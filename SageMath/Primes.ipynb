```python
class PrimeList(list):
    def __init__(self):
        self = []
    def generate(self, n):
        del self[:]
        for num in range(2,n+1):
            if is_prime(num):
                self.append(num)
        return self
```

```python
a = PrimeList(); 
%time len(a.generate(1000))
```

```python
N = 10000; B = 1.08366; switch = 0;

plist = a.generate(N); plist
datalist = [(k,len([x for x in plist if x <= k])) for k in range(3,N+1,int(N/100))]
G1 = list_plot(datalist, pointsize = 1)

Li(x) = x/(log(x)-B*switch)
G2 = plot(Li,(x,3,N), color = "red")

var('u,t'); assume(t-2>0)
GI(t) = integral(1/log(u),(u,2,t)) #Ei(log(x)) #
G3 = plot(GI,(x,3,N), color = "green")

(G1+G2+G3).show(ymin=0,ymax=max([n[1] for n in datalist])*1.2)
```
