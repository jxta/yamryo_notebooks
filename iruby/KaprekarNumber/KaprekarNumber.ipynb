# Kaprekar Numbers

## Define Kaprekar Class

```ruby
module Kaprekar_interface
  def initialize(num = 1)
    @number = num
  end
  
  def get_next() end
  def sequence() end
  def find_loop() end
  
  def ==(kp) @number == kp.number end
  def inspect() @number end

  private
  def split(num)
    raise ArgumentError unless num.is_a?(Integer)
    return num.to_s.split("")
  end
  def min(num) split(num).sort.join.to_i end
  def max(num) split(num).sort.reverse.join.to_i end
end

class Kaprekar include Kaprekar_interface
  def initialize(num)
    super(num)
  end
  attr_reader :number
  
  def get_next() 
    num = (max(@number) - min(@number))
    self.class.new(num) 
  end
  
  def sequence()
    seq = [self]
    while (seq.count(seq[-1]) < 2)
      seq << seq[-1].get_next
    end
    return seq.drop(1)
  end

  def find_loop()
    seq = self.sequence
    idx = seq.index(seq[-1])
    return seq[idx, seq.length - idx]
  end
end; nil
```

```ruby
nn = 10**3
(nn..nn*10).each_with_object([]) do |n, a|
  kp = Kaprekar.new(n)
  mystr = kp.number.to_s
  while !a.include?(kp)
    a << kp
    kp = kp.get_next
    mystr += '->'+ kp.number.to_s
  end
  p mystr
end; nil
```

## Display Kaprekar Tree

```ruby
Kplist = ((100..1000).to_a).map!{|n| Kaprekar.new(n)}
Nodes = Kplist.map{|kp| [kp, Kplist.select{|x| x.get_next == kp}]}.to_h

def maketree(kp, maxdepth)
  children = Nodes[kp]
  subtree = (children.empty? or maxdepth == 0) ? nil : children.map{|c| maketree(c, maxdepth-1)}
  return {kp => subtree}
end

def display(tree, stack=[], str='')
  dp = str.length
  tree.each do |root, subtree|
    unless stack.include?(root) then
      puts str + root.number.to_s
      stack << root
      unless subtree.nil? then
        subtree.each do |c|
          display(c, stack, ' '*(dp+1)+'|-')
        end
      end
    else
      puts str + '(' + root.number.to_s + ')'
    end
  end
  return stack
end; nil
```

```ruby
stack = []
Kplist.each do |kp|
  unless Nodes[kp].empty? then
    tree = maketree(kp, 10)
    display(tree, stack)
  end
end; nil
```
