# Distribution of digits in the decimal representation of the square root of primes

```python
import matplotlib.pyplot as plt
%matplotlib inline

from decimal import *
N = 100
getcontext().prec = N

num_list = [2,3,5,7,11,13,17,19,23] # range(2,11) # 
sqrt_list = [str(int(Decimal(n).sqrt()*(10**N))) for n in num_list]
# sqrt_list = [str(Decimal(n).sqrt()).replace('.','') for n in num_list]
data_list = [{str(c): a.count(str(c)) for c in range(10)} for a in sqrt_list]
# data_list[1]
```

```python
plt.figure(figsize=(16,10))

for i in range(len(num_list)):
    plt.subplot(331+i)
    X = data_list[i].keys()
    Y = data_list[i].values()
    plt.bar(range(10), Y, facecolor='#9999ff', edgecolor='white')
    plt.xticks(range(10),X)
    plt.title('sqrt of {}'.format(num_list[i]))
    
plt.show()
```
