```python
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
%config inlineBackend.format_figure = 'retina'
```

```python
N = 100*2

xlist = np.array([k/N for k in range(N)], dtype='float64')
sqrtlist = np.array([np.sqrt(x) for x in xlist], dtype='float64')
T = np.array([sqrtlist[k]-xlist[k] for k in range(N)], dtype='float64')

plt.figure(figsize=[8,8])
plt.grid(True)                

plt.plot(xlist, sqrtlist)
plt.plot(xlist, xlist)
plt.plot(xlist, T)
```

```python
pnts = [np.array([xlist[k], T[k]]) - np.array([1/16, 3/16]) for k in range(N)]

xx = [p[0] for p in pnts]
yy = [p[1] for p in pnts]

plt.figure(figsize=[8,8])

plt.plot(xlist, xlist)
plt.plot(xlist, T)
plt.plot(xx, yy)
```

```python
def Rot0(_rad):
    return np.array([
        [np.cos(_rad), -np.sin(_rad)],
        [np.sin(_rad), np.cos(_rad)],
    ])
```

```python
M = Rot0(np.deg2rad(-45))

vecs = [ M.dot(p) for p in pnts]
#vecs = [np.array([xlist[k], T[k]]) for k in range(N)]

xx = [v[0] for v in vecs]
yy = [v[1] for v in vecs]
sq = [(-2.83)*x**2 for x in xx]

plt.figure(figsize=[8,8])
plt.axis([-0.2, 1.0, -.9, 0.3])

#plt.plot(xlist, xlist)
plt.plot(xlist, T)
plt.plot(xx, yy)
plt.plot(xx, sq)
None
```
