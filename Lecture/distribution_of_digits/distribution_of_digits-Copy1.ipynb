# Distribution of digits in the decimal representation of the square root of primes

```python
from decimal import *

def get_hst(num, order=100):
    getcontext().prec = order
    sqrt_str = str(int(Decimal(num).sqrt()*(10**order)))
    return {str(c): sqrt_str.count(str(c)) for c in range(10)}
```

```python
print(get_hst(2, 10000))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(16,12))

num = 3

ods = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]
for i in range(len(ods)):
    plt.subplot(331+i)
    X,Y = [],[]
    for key, value in sorted(get_hst(num,ods[i]).items()):
        X.append(key)
        Y.append(value)
    plt.bar(range(10), Y, facecolor='#9999ff', edgecolor='white')
    plt.xticks(range(10), X)
    plt.title('order = {}'.format(ods[i]))
    
plt.show()
```

```python
plt.figure(figsize=(16,12))

order = 10000

nums = [2, 3, 5, 7, 11, 13, 17, 19, 23]
for i in range(len(nums)):
    plt.subplot(331+i)
    X,Y = [],[]
    for key, value in sorted(get_hst(nums[i],order).items()):
        X.append(key)
        Y.append(value)
    plt.bar(range(10), Y, facecolor='#9999ff', edgecolor='white')
    plt.xticks(range(10), X)
    plt.title('order = {}'.format(nums[i]))
    
plt.show()
```
