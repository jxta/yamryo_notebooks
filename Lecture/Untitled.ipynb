```python
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
mm, MM = -5, 5

bp_x = np.linspace(mm, MM, num=40, endpoint=True)
bp_y = np.exp(bp_x)

# The graph of the given function
plt.plot(bp_x, bp_y, linewidth=2, linestyle="--",
         color="blue", label=r"$\sin(x)$")
# A graph of a polynomial
poly_y = [1+1*t+2*t**2 for t in bp_x]
plt.plot(bp_x, poly_y, linewidth=1.2, linestyle="-", color="green", label=r"polynomial")

# Drawing
plt.title(r"Taylor expansion experiments")
plt.xlim(mm, MM)
plt.ylim(-10, 20)
plt.legend(loc="lower left")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Silly example data
X = np.arange(0.25, 2*np.pi, 0.25)
Y = np.arange(0.25, 2*np.pi, 0.25)
X, Y = np.meshgrid(X, Y)
R = X/(2*Y)
Z = np.arccos(R)

# Make the plot
fig = plt.figure()
ax = fig.gca(projection="3d")
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(0, np.pi)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
