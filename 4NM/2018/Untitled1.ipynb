```python
import math, cmath
import numpy as np
from tqdm import tqdm_notebook as tqdm

import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.path as mpath
import matplotlib.patches as mpatches
```

```python
z = 1j; a = 1+(1/1)*1j
seq = [int(z.real)]

N = 10**5
for k in tqdm(range(N)):
    z = (z-1)/(z+a)
#    z = 1/(a*z).conjugate()
    seq.append(z)
```

```python
plt.figure(figsize=(18,10))

Path = mpath.Path
N = 10**2
path_data = [(seq[i].real,seq[i].imag) for i in range(0,N+1,1)]

path = mpath.Path([list(c) for c in path_data])
x, y = zip(*path.vertices)
line, = plt.plot(x, y, 'go-', ms = 2.0, lw = 0.2)

plt.grid()
plt.axis('equal')
# plt.xscale('symlog'); plt.yscale('symlog')

plt.show()
```

```python
plt.figure(figsize=(16,16))

d = 4

plt.subplot(211)
N = d*10
for f in range(4):
    path_data = [(seq[i].real,seq[i].imag) for i in range(f,N+1,d)]
    path = mpath.Path([list(c) for c in path_data])
    x, y = zip(*path.vertices)
    line, = plt.plot(x, y, 'o-', ms = 2.0, lw = 0.2)
    plt.grid(True)

plt.subplot(212)
N = d*5
for f in range(d):
    fn = (lambda n: f+d*n)
    X = range(f,N+1+f)
    Y = [seq[fn(k)].imag for k in X]
    plt.plot(X, Y, '.-')
    plt.grid(True)

plt.show()
```

```python
N = 80
plt.figure(figsize=(16,16))
for s in range(1,21):
    plt.subplot(5, 4, 1+(s-1))
    plt.plot(range(0,N+1,s), [z.real for z in seq[0:N+1:s]])
    plt.grid(True)
    plt.title('step={}'.format(s))

plt.show()
```
