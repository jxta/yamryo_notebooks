```python
import math,cmath

# from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cmath.polar(1+3j)
```

```python
n = 100

rlist = [k/(k+1) for k in range(1,n)]
alist = [math.pi/k/(k+1) for k in range(1,n)]

zlist = [rlist[k]*(math.cos(alist[k])+1j*math.sin(alist[k])) for k in range(n-1)]
```

```python
import matplotlib.path as mpath
import matplotlib.patches as mpatches

plt.figure(figsize=(16,16))

Path = mpath.Path

path_data = [(z.real, z.imag) for z in zlist]
path = mpath.Path([list(c) for c in path_data])
x, y = zip(*path.vertices)
line, = plt.plot(x, y, 'go-')

plt.grid()
plt.axis('equal')
plt.show()
```

```python
import matplotlib.path as mpath
import matplotlib.patches as mpatches

plt.figure(figsize=(16,16))

Path = mpath.Path

path_data = []
w = 1
for k in range(n-1):
    w = w*zlist[k]
    path_data.append(((k*w).real, (k*w).imag))

path = mpath.Path([list(c) for c in path_data])
x, y = zip(*path.vertices)
line, = plt.plot(x, y, 'go-')

plt.grid()
plt.axis('equal')

plt.show()
```
