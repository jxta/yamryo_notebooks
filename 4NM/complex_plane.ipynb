```python
z = 1j
a = 1+1j
seq = [int(z.real)]
N = 10**4
for k in range(N):
    z *= a
    z = 1/z.conjugate()
    seq.append(z)
print(seq[0:6])
print([1,
       (a/abs(a))/abs(a), #1/a.conjugate(),
       (a/abs(a))**2, # abs(a)**2/a.conjugate()**3, 
       (a/abs(a))**3/abs(a), #(a**3)/abs(a)**4, 
       (a/abs(a))**4, #(a**4)/abs(a)**4, # abs(a)**4/a.conjugate()**4,
       (a/abs(a))**5/abs(a), #(a**5)/abs(a)**6, # abs(a)**4/a.conjugate()**5,
      ])
```

```python
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
N = 400
plt.figure(figsize=(16,10))
for s in range(1,17):
    plt.subplot(4, 4, 1+(s-1))
    plt.plot(range(0,N+1,s), seq[0:N+1:s])
    plt.grid(True)
    plt.title('step={}'.format(s))

plt.show()
```

```python
import matplotlib.path as mpath
import matplotlib.patches as mpatches

plt.figure(figsize=(16,16))

Path = mpath.Path
N = 800
# path_data = [(seq[i].real,seq[i].imag) for i in range(0,N+1,10)]
path_data = []
for k in range(int(N/8)):
    z = (a/abs(a))**(10*k)+(a/abs(a))**(10*k+5/2)
    path_data.append((z.real, z.imag))

path = mpath.Path([list(c) for c in path_data])
x, y = zip(*path.vertices)
line, = plt.plot(x, y, 'go-')

plt.grid()
plt.axis('equal')

plt.show()
```
